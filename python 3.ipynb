````python


---

### **1. What is the difference between a function and a method in Python?**

* A **function** is a block of reusable code defined using `def` and can be called independently.
* A **method** is a function that is associated with an object (i.e., it belongs to a class).

**Example:**

```python
def my_function():
    print("I am a function")

class MyClass:
    def my_method(self):
        print("I am a method")
```

👉 Key difference:

* Function → standalone
* Method → tied to an object (uses `self`)

---

### **2. Explain the concept of function arguments and parameters in Python**

* **Parameters** are variables defined in the function declaration.
* **Arguments** are the actual values passed when calling the function.

**Example:**

```python
def greet(name):  # 'name' is a parameter
    print("Hello", name)

greet("Aman")     # "Aman" is an argument
```

👉 Types of arguments:

* Positional
* Keyword
* Default
* Variable-length (`*args`, `**kwargs`)

---

### **3. What are the different ways to define and call a function in Python?**

**Ways to define:**

1. Regular function (`def`)
2. Lambda function (`lambda`)
3. Nested function
4. Recursive function

**Ways to call:**

```python
def add(a, b):
    return a + b

add(2, 3)                # positional
add(a=2, b=3)            # keyword
add(2, b=3)              # mixed
```

---

### **4. What is the purpose of the `return` statement in a Python function?**

* It sends a result back to the caller.
* It also ends the function execution.

**Example:**

```python
def square(x):
    return x * x
```

👉 Without `return`, the function returns `None` by default.

---

### **5. What are iterators in Python and how do they differ from iterables?**

* **Iterable**: An object that can be looped over (e.g., list, tuple, string).
* **Iterator**: An object that keeps track of iteration using `__next__()`.

**Example:**

```python
my_list = [1, 2, 3]       # iterable
it = iter(my_list)        # iterator

print(next(it))           # 1
```

👉 Difference:

* Iterable → can be converted to iterator
* Iterator → produces values one at a time

---

### **6. Explain the concept of generators in Python and how they are defined**

* Generators are functions that yield values one at a time using `yield`.

**Example:**

```python
def count_up(n):
    for i in range(n):
        yield i
```

👉 They don't store all values in memory at once.

---

### **7. What are the advantages of using generators over regular functions?**

* Memory efficient (lazy evaluation)
* Faster for large datasets
* Produce values on demand

👉 Example: reading large files line-by-line instead of loading entire file.

---

### **8. What is a lambda function in Python and when is it typically used?**

* A **lambda function** is a small anonymous function defined using `lambda`.

**Example:**

```python
square = lambda x: x * x
print(square(5))
```

👉 Used when:

* Short, simple operations are needed
* Often used with `map()`, `filter()`, etc.

---

### **9. Explain the purpose and usage of the `map()` function in Python**

* `map()` applies a function to all items in an iterable.

**Example:**

```python
nums = [1, 2, 3]
result = map(lambda x: x * 2, nums)
print(list(result))   # [2, 4, 6]
```

👉 Useful for transforming data.

---

### **10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

| Function   | Purpose                            |
| ---------- | ---------------------------------- |
| `map()`    | Applies function to each item      |
| `filter()` | Filters items based on condition   |
| `reduce()` | Reduces iterable to a single value |

**Examples:**

```python
from functools import reduce

nums = [1, 2, 3, 4]

# map
print(list(map(lambda x: x*2, nums)))

# filter
print(list(filter(lambda x: x % 2 == 0, nums)))

# reduce
print(reduce(lambda x, y: x + y, nums))
```

---

````

````python
#PRACTICLE

---

### **1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list**

```python
def sum_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

# Example
print(sum_even_numbers([47, 11, 42, 13]))  # Output: 42
```

---

### **2. Create a Python function that accepts a string and returns the reverse of that string**

```python
def reverse_string(text):
    return text[::-1]

# Example
print(reverse_string("hello"))  # Output: olleh
```

---

### **3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number**

```python
def square_list(numbers):
    return [num**2 for num in numbers]

# Example
print(square_list([1, 2, 3, 4]))  # Output: [1, 4, 9, 16]
```

---

### **4. Write a Python function that checks if a given number is prime or not (1 to 200)**

```python
def is_prime(n):
    if n <= 1 or n > 200:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example
print(is_prime(47))  # Output: True
```

---

### **5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms**

```python
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        value = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return value

# Example
for num in Fibonacci(5):
    print(num)  # Output: 0 1 1 2 3
```

---

### **6. Write a generator function in Python that yields the powers of 2 up to a given exponent**

```python
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

# Example
for val in powers_of_two(5):
    print(val)  # Output: 1 2 4 8 16 32
```

---

### **7. Implement a generator function that reads a file line by line and yields each line as a string**

```python
def read_file_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()

# Example
for line in read_file_lines("sample.txt"):
    print(line)
```

---

### **8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple**

```python
data = [(1, 3), (4, 1), (2, 2)]

sorted_data = sorted(data, key=lambda x: x[1])

print(sorted_data)  # Output: [(4, 1), (2, 2), (1, 3)]
```

---

### **9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit**

```python
celsius = [0, 10, 20, 30]

fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print(fahrenheit)  # Output: [32.0, 50.0, 68.0, 86.0]
```

---

### **10. Create a Python program that uses `filter()` to remove all the vowels from a given string**

```python
def remove_vowels(text):
    vowels = "aeiouAEIOU"
    result = ''.join(filter(lambda x: x not in vowels, text))
    return result

# Example
print(remove_vowels("Hello World"))  # Output: Hll Wrld
```

---
````
